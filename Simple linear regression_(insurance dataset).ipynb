```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
insurance = pd.read_csv("insurance.csv")

# get the number of missing data points per column
missing_values_count = insurance.isnull().sum()

# look at the # of missing points in the first ten columns
missing_values_count[:]

insurance.dtypes
```

```python
insurance.head()
```

```python
#Scatter plot of age vs charges  
x = insurance.iloc[:,0].values
y = insurance.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#Line of best fit on Scatter plot of age vs charges 
insurance_model = LinearRegression()
x=x.reshape(-1,1)
insurance_model.fit(x,y)

#Making prediction
y_pred = insurance_model.predict(x)
unk_x = [[40]]

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
print(f"{unk_x} year old should pay:, {insurance_model.predict(unk_x)}")
```

```python
#Scatter plot of BMI vs charges  
x = insurance.iloc[:,2].values
y = insurance.iloc[:,-1].values
#Fitting linear regression model
insurance_model_ = LinearRegression()
x=x.reshape(-1,1)
insurance_model_.fit(x,y)

y_pred = insurance_model_.predict(x)
unk_x = [[40]]

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.show()
print(f"A person with a BMI of {unk_x} should pay:, {insurance_model_.predict(unk_x)}")
```
